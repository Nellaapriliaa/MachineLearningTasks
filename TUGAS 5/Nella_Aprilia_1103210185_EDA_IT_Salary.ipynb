ChatGPT Prompt: beri mode untuk menampilkan pie diagram chart untuk fitur age dengan min 20, max 69

```python
# @title 5. Visualisasikan Data

import matplotlib.pyplot as plt

# Menentukan rentang nilai umur
min_age = 20
max_age = 69


# Mengelompokkan data umur ke dalam kategori
bins = [min_age, 30, 40, 50, 60, max_age]
labels = ['20-30', '31-40', '41-50', '51-60', '61+']

# Mengelompokkan data umur ke dalam kategori
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Menghitung jumlah data dalam setiap kategori
age_counts = df['Age_group'].value_counts()

# Membuat pie chart untuk umut
plt.figure(figsize=(8, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Usia (20-69 tahun)')
plt.axis('equal')  # Membuat lingkaran menjadi sejati
plt.show()
```

```python
import matplotlib.pyplot as plt

# Memberikan nilai numerik kepada setiap kategori gender
gender_mapping = {'Male': 0, 'Female': 1}
df['Gender_Code'] = df['Gender'].map(gender_mapping)

# Membuat histogram untuk kolom gender yang sudah dikodekan
plt.hist(df['Gender_Code'], bins=2, color='skyblue', edgecolor='black')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Histogram of Gender')
plt.xticks(range(2), ['Male', 'Female'])
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Menghitung korelasi antar fitur numerik dalam DataFrame
correlation_matrix = df.corr()

# Membuat heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()
```
