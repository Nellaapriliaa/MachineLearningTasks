```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/Economy_Indicators.csv')

# Memilih hanya fitur-fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan boxplot untuk setiap fitur numerik
for column in numerical_features.columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# @title 6. Distribution Plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV ke dalam DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/Economy_Indicators.csv')

# Memilih hanya kolom numerik
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Menentukan ukuran plot
plt.figure(figsize=(14, 10))

# Melakukan loop untuk setiap fitur numerik dan membuat distribution plot
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# @title 7. Violin Plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV ke dalam DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/Economy_Indicators.csv')

# Memilih hanya kolom numerik
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Menentukan ukuran plot
plt.figure(figsize=(14, 10))

# Melakukan loop untuk setiap fitur numerik dan membuat violin plot
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(3, 3, i + 1)
    sns.violinplot(data=df, y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# @title 8. Heat Map
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv('/content/drive/MyDrive/Dataset/Economy_Indicators.csv')

# Memilih hanya fitur numerik
numeric_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numeric_features.corr()

# Membuat heat map
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
